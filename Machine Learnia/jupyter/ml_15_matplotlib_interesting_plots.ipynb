```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()

x = iris.data
y = iris.target

# display(x)
# display(y)
names = iris.target_names
print(names)
print(f'x contient {x.shape[0]} exemples et {x.shape[1]} variables')
print(f'il y a {np.unique(y).size} classes')

plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5, s=x[:, 2] * 100)
plt.xlabel('longueur sépal')
plt.ylabel('largeur sépal')
```

```python
%matplotlib
from mpl_toolkits.mplot3d import Axes3D
ax = plt.axes(projection='3d')
plt.scatter(x[:, 0], x[:, 1], x[:, 2], c=y)
```

```python
f = lambda x, y: np.sin(x) + np.cos(x + y)
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)
Z.shape
plt.figure(figsize=(16, 9))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')
```

```python
plt.figure(figsize=(16, 9))
plt.hist(x[:, 0], bins=30)
plt.hist(x[:, 1], bins=30)
```

```python
plt.figure(figsize=(16, 9))
plt.hist2d(x[:, 0], x[:, 1], bins=30, cmap='Blues')
plt.xlabel('longueur sépal')
plt.ylabel('largeur sépal')
plt.colorbar()
```

```python
from scipy import misc
face = misc.face(gray=True)
# plt.imshow(face, cmap='gray')
plt.figure(figsize=(16, 9))
plt.hist(face.ravel(), bins=255)
plt.show()
```

```python
plt.figure(figsize=(16, 9))
plt.contour(X, Y, Z, 50) #, colors='black')
```

```python
plt.figure(figsize=(16, 9))
plt.contourf(X, Y, Z, 20, cmap='plasma')
plt.colorbar()
```

```python
plt.figure(figsize=(16, 9))
plt.imshow(face)
```

```python
plt.figure(figsize=(16, 9))
plt.imshow(np.corrcoef(x.T))
plt.colorbar()
```
