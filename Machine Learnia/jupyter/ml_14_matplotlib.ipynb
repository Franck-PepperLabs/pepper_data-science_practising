## 14 / 30 Matplotlib

Eviter de mélanger l'approche OO et Func.

```python
import numpy as np
dataset = {f'experience {i}': np.random.randn(100) for i in range(4)}
print(dataset)
```

```python
import matplotlib.pyplot as plt
def graphique(dataset):
  x = range(100)
  plt.figure(figsize=(16, 9))
  for i in range(4):
    plt.subplot(4, 1, i + 1)
    plt.plot(x, dataset[f'experience {i}'])
    plt.title(f'experience {i}')
  plt.show()
graphique(dataset)
```
