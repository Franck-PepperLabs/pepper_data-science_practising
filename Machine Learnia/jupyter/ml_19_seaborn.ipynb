```python
titanic = sns.load_dataset('titanic')
display(titanic.head(3))
```

```python
titanic.drop(['alone', 'alive', 'who', 'adult_male', 'embark_town', 'class'], axis=1, inplace=True)
titanic.dropna(axis=0, inplace=True)
display(titanic.head(3))
```

```python
sns.pairplot(titanic)
```

## Catégories

Les alignements ci-dessus sont la conséquence des données catégorielles.

```python
sns.catplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sns.boxplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sns.distplot(titanic.fare) # https://gist.github.com/mwaskom/de44147ed2974457ad6372750bbe5751
plt.show()
sns.histplot(titanic.fare, kde=True, stat='density', kde_kws=dict(cut=3))
plt.show()
```

```python
sns.jointplot(data=titanic, x='age', y='fare')
sns.jointplot(data=titanic, x='age', y='fare', kind='kde')
sns.jointplot(data=titanic, x='age', y='fare', kind='hex')
```

```python
sns.heatmap(titanic.corr())
```
