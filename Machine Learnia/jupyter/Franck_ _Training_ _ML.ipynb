### Résolution algébrique de thêta

$\hat{y}=x^T\theta = \theta_0x_0+\theta_1x_1+...+\theta_{N-1}x_{N-1}$ avec :
- $x^T=\left(x_1, x_2, x_3, ..., x_N\right)$
- $\hat{y}^T=\left(\hat{y}_1, \hat{y}_2, \hat{y}_3, ..., \hat{y}_N\right)$

$x^T_i=\left(1, v_{i, 1}, v_{i, 2}, ..., v^i_{i, D}\right)$, avec $D$ nombre de variables qui caractérisent l'observation.

Ici il n'y a qu'une variable surface, $v_{i, 1}$ soit $x^T_i=\left(1, v_i\right)$.

$\theta=\left(\theta_0, \theta_1\right)$


$\hat{y}=\begin{pmatrix}
\hat{y}_1\\
...\\
\hat{y}_N
\end{pmatrix}=\begin{pmatrix}
x^T_1=\left(1, v_{1, 1}\right)\\
...\\
x^T_N=\left(1, v_{1, N}\right)
\end{pmatrix}\begin{pmatrix}
\theta_0\\
\theta_1
\end{pmatrix}=x^T\theta$

bruit irréductible, erreur : distance entre la périvision et l'observation réelle :

$d_2\left(y, \hat{y}\right)=\left\lVert\hat{y}-y\right\rVert_2=\left(d_2\left(y_i, \hat{y}_i\right)=\left\lVert\hat{y}_i-y_i\right\rVert_2\right)_{i \in \left[1, .., N\right]}$

$\left\lVert.\right\rVert_2$ représente la norme euclidienne.

Sur un espace d'observation de dimension $d$ : $d_2\left(y_i, \hat{y}_i\right)=\left\lVert\hat{y}_i-y_i\right\rVert_2=\sqrt{\sum\limits_{j=1}^d{\left(\hat{y}_{ij}-y_{ij}\right)^2}}$.

Le risque empirique : $E=\sum^N_{i=1} \left\lVert\hat{y}_i-y_i\right\rVert_2$

Et donc : $E=\sum\limits_{i=1}^n{d_2\left(y_i, \hat{y}_i\right)}=\sum\limits_{i=1}^n{d_2\left(y_i, x^T_i\theta\right)}=\sum\limits_{i=1}^n{\sqrt{\sum\limits_{i=1}^d{\left(x^T_i\theta-y_i\right)^2}}}$

On cherche le $\theta$, noté $\hat{\theta}$ qui minimine $E_\theta$ : $\operatorname{arg\,min}_\theta E$

En partant d'ici, c'est plus clair : https://eli.thegreenplace.net/2014/derivation-of-the-normal-equation-for-linear-regression


Fontion d'hypothèse : $h_g\left(x\right)=\theta_0x_0+\theta_1x_1+...+\theta_nx_n$

En posant $\theta=\begin{pmatrix}
\theta_0\\
...\\
\theta_n
\end{pmatrix}\in \mathbb{R}^{n+1}$

Réécriture : $h_g\left(x\right)=\theta^Tx$

```python
# On décompose le dataset et on le transforme en matrices pour pouvoir effectuer notre calcul
N = house_data.shape[0]   # Nombre d'observations (lignes)
D = 1                     # Nombre de variables par observation (colonnes)
X = np.matrix([np.ones(N), house_data['surface'].values]).T
display(X)

y = np.matrix(house_data['loyer']).T
#display(y)

# On effectue le calcul exact du paramètre theta
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
inv_X = np.linalg.inv(X.T.dot(X))
# display(inv_X) l'intérêt, c'est que xTx est carrée alors que x est grande et pas carrée
#theta = np.linalg.inv(X).dot(y)

print(theta)

# https://eli.thegreenplace.net/2014/derivation-of-the-normal-equation-for-linear-regression
```

### Sampling

Drôle de conseil : un sample avec des index répétés (il refait random.choices, alors qu'il devrait utiliser rndom.sample(seq, n)

```python
import random

def plot_sample_indexes(sample, sample_max, title=None, xlabel=None, ylabel=None) :
  fig, ax = plt.subplots(1)  #, figsize=(16, 9))

  # On affiche le nuage de points dont on dispose
  ax.set_xlabel(xlabel)
  if not ylabel is None:
    ax.set_ylabel(ylabel)
  ax.grid(True)
  ax.set_title(title, fontsize=20)
  ax.plot(sample)

  # On affiche la droite entre 0 et 250
  ax.plot([0, len(sample)], [0, sample_max], linestyle='--', c='#000000')
  plt.show()


def plot_counts_frequencies(v, title=None, xlabel=None, ylabel=None):
  i = min(v)
  a = max(v)
  n = len(set(v))
  print('min :', i, 'max :', a, 'max - min = ', a - i, 'n_uniques :', n)
  x = np.arange(i, a + 1)
  # print(x)
  s = pd.Series(v)
  u = s.value_counts()
  #display(u)
  y = []
  # y = [u.iloc[x_i, 1] for x_i in x]
  for x_i in x:
    if not x_i in u.index:
      y.append(0)
    else:
      y.append(u[x_i])
  #display(y)

  fig, ax = plt.subplots(1)#, figsize=(16, 9))

  ax.set_xlabel(xlabel)
  if not ylabel is None:
    ax.set_ylabel(ylabel)
  ax.grid(True)
  ax.set_title(title, fontsize=20)
  
  ax.bar(x, y)
  #ax.plot(x, y)

  # On affiche la droite entre 0 et 250
  # ax.plot([0, len(sample)], [0, sampled_size], linestyle='--', c='#000000')
  plt.show()



# sz = len(house_data.index)
# print(sz)

```

### Quelle fréquence par longueur de saut ?

```python

source_sz = 10000
ar = np.arange(0, source_sz)

sz = sample_size = source_sz // 11

sample_a = np.random.randint(source_sz, size=sz)
sample_a.sort()

if sz < 100:
  print('sample_a[1:' + str(len(sample_a)) + ']', sample_a)

sample_jumps_a = sample_a[1:] - sample_a[:-1]

if sz < 100:
  print('jumps[1:' +  str(sample_a.size) + ']', sample_jumps_a)

plot_sample_indexes(sample_a, source_sz, title='np.random.randint', xlabel='N° Tirage', ylabel='Index tiré')

plot_counts_frequencies(
    sample_jumps_a,
    title='Taille du saut entre deux index successifs',
    xlabel='Taille du saut', ylabel='Fréquence')


sample_b = np.array(random.sample(ar.tolist(), sz))
sample_b.sort()

if sz < 100:
  print('sample_a[1:' + str(len(sample_b)) + ']', sample_b)
sample_jumps_b = sample_b[1:] - sample_b[:-1]

if sz < 100:
  print('jumps[1:' +  str(sample_b.size) + ']', sample_jumps_b)

plot_sample_indexes(sample_b, source_sz, title='random.sample', xlabel='N° Tirage', ylabel='Index tiré')

plot_counts_frequencies(
    sample_jumps_b,
    title='Taille du saut entre deux index successifs',
    xlabel='Taille du saut', ylabel='Fréquence')

```

```python
def print_mnist_sample(index, data, target, m=7, n=4):
  fig, axes = None, None
  if len(index) > n:
    m = min(m, len(index) // n)
    fig, axes = plt.subplots(m, n, figsize=(20, 35))  #, figsize=(16, 9))
  else:
    m = 1
    n = len(index)
    fig, axes = plt.subplots(m, n, figsize=(20, 35))  #, figsize=(16, 9))

  # flatten axes for a one loop enumeration
  axes = axes.reshape(m * n)
  for i, ax in zip(index, axes):
    ax.grid(True)
    ax.set_xlabel('Mnist N° ' + str(i))
    ax.set_title(target[i], fontsize=20)
    ax.imshow(data[i].reshape(28, 28))

  plt.show()


data = mnist.data.values
target = mnist.target.values
print_mnist_sample([3, 5, 7, 11], data, target, m=7, n=5)
print_mnist_sample(range(50, 62), data, target, m=7, n=4)
```
