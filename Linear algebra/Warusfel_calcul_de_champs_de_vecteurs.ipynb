## 1. Gradient, divergence, rotationnel

### 1.1 Différentielle, matrice jacobienne

Soit $\vec{V}$ un champ de vecteurs de classe $\mathcal{C}^1$ sur $U$.

<mark>Si je me souviens de la signification de la classe $\mathcal{C}^1$ (une fois dérivable), c'est également une notion dont je dois parfaitement m'assurer en contexte de ces révisions des fondamentaux</mark>

Ses fonctions composantes $(V_i)_{1 \le i \le n}$ sont alors de classe $\mathcal{C}^1$ sur $U$ et vérifient, pour tout $M \in U$ et $\vec{h} = (h_1, \cdots, h_n)$ tel que $M + \vec{h} \in U$ :

$\begin{align}
    V_i(M + \vec{h}) = V_i(M) +
        \displaystyle\sum^n_{j = 1} \frac{\partial V_i}{\partial x_j}(M)h_j +
        \|\vec{h}\|\varepsilon_i(\vec{h})
    \tag{1}
\end{align}$

avec $\displaystyle\lim_0{\varepsilon_i} = 0$.

Pour tout point de $M \in U$, il existe donc une *application linéaire* notée $d\vec{V}_M$ telle que :

$\begin{align}
    \vec{V}(M + \vec{h}) = \vec{V}(M) +
        d\vec{V}_M(\vec{h}) +
        \|\vec{h}\|\varepsilon(\vec{h})
    \tag{2}
\end{align}$

avec $\displaystyle\lim_0{\varepsilon} = 0$.

Cette application linéaire, appelée *différentielle* de $\vec{V}$ en $M$, a pour matrice dans la base canonique de $\mathbb{R}^n$ :

$\begin{align}
    J = \left( \frac{\partial V_i}{\partial x_j}(M) \right)_{1 \le i, j \le n}
    \tag{3}
\end{align}$

appelée *matrice jacobienne* de $\vec{V}$ en $M$.

### 1.1 Gradient

#### Rappels

Soit $f$ une application de classe $\mathcal{C}^1$ sur $U$ à valeurs dans $\mathbb{R}$. Pour $M \in U$ et $\vec{h} = (h_1, \cdots, h_n)$ tel que $M + \vec{h} \in U$, on a :

$\begin{align}
    f(M + \vec{h}) = f(M) +
        \displaystyle\sum^n_{j = 1} \frac{\partial f}{\partial x_j}(M)h_j +
        \|\vec{h}\|\varepsilon(\vec{h})
    \tag{4}
\end{align}$

avec $\displaystyle\lim_0{\varepsilon} = 0$.

Le vecteur $\overrightarrow{\text{grad}}_M f$ dont les composantes dans la basse canonique de $\mathbb{R}^n$ sont $\left( \frac{\partial f}{\partial x_j}(M) \right)_{1 \le i \le n}$ est appelé *gradient* de $f$ en $M$. Il est tel que :

$\begin{align}
    f(M + \vec{h}) = f(M) +
        \vec{h}.\overrightarrow{\text{grad}}_M f +
        \|\vec{h}\|\varepsilon(\vec{h})
    \tag{5}
\end{align}$

avec $\displaystyle\lim_0{\varepsilon} = 0$.

On note $\overrightarrow{\text{grad}} f$ l'application $M \mapsto \overrightarrow{\text{grad}}_M f$.

#### Propriétés

* L'application $f \mapsto \overrightarrow{\text{grad}} f$ est un application linéaire de $\mathcal{C}^1 (U, \mathbb{R})$ dans $\mathcal{C}^0 (U, \mathbb{R})$.
* Elle vérifie, pour $f \in \mathcal{C}^1 (U, \mathbb{R})$ et $g \in \mathcal{C}^1 (U, \mathbb{R})$ :

$\begin{align}
    \overrightarrow{\text{grad}} (f \, g) =
        f \, \overrightarrow{\text{grad}} \, g +
        g \, \overrightarrow{\text{grad}} \, f
    \tag{6}
\end{align}$

*Exemples*

On a vu page 623 que le gradient d'une fonction $f$ définie sur $U = \mathbb{R}^n \setminus \{0\}$ par :

$\begin{align}
    f(M) = \varphi(r)
    \tag{a}
\end{align}$

où $r = \|\overrightarrow{OM}\|$ est :

$\begin{align}
    \overrightarrow{\text{grad}}_M \, f = \varphi' (r) \frac{\overrightarrow{OM}}{\|\overrightarrow{OM}\|}
    \tag{b}
\end{align}$

En particulier, si :

$\begin{align}
    f(M) = \frac{q}{4 \pi \varepsilon_0 r}
    \tag{c}
\end{align}$

alors :

$\begin{align}
    \overrightarrow{\text{grad}}_M \, f =
    -\frac{q}{4 \pi \varepsilon_0 r} \frac{\overrightarrow{OM}}{\|\overrightarrow{OM}\|^3} =
    -\frac{q}{4 \pi \varepsilon_0 r^2} \vec{u}
    \tag{d}
\end{align}$

où $\vec{u} = \frac{\overrightarrow{OM}}{\|\overrightarrow{OM}\|}$.

## 1.3 Divergence

On appelle *divergence* d'un champ de vecteurs $\vec{V}$ de classe $\mathcal{C}^1$ sur $U$ la fonction définie sur $U$ par :

$\begin{align}
    \text{div}_M \, \vec{V} = \displaystyle\sum^n_{i = 1} \frac{\partial V_i}{\partial x_i}(M)
    \tag{7}
\end{align}$

On note $\text{div} \, \vec{V}$ l'application $M \mapsto \text{div}_M \, \vec{V}$.

#### Propriétés

* L'application $\vec{V} \mapsto \text{div} \, \vec{V}$ est un application linéaire de $\mathcal{C}^1 (U, \mathbb{R}^n)$ dans $\mathcal{C}^0 (U, \mathbb{R}^n)$.
* Elle vérifie, pour $f \in \mathcal{C}^1 (U, \mathbb{R})$ et $\vec{V} \in \mathcal{C}^1 (U, \mathbb{R}^n)$ :

$\begin{align}
    \text{div} (f \, \vec{V}) = \left( \overrightarrow{\text{grad}} \, f \right).\vec{V} + f \, \text{div} \, \vec{V}
    \tag{8}
\end{align}$

*Exemples*

Si $\vec{V}(M) = \displaystyle\frac{\overrightarrow{OM}}{\|\overrightarrow{OM}\|^\alpha}$, on a $V_i(M) = \displaystyle\frac{x_i}{r^\alpha}$ et donc :

$\begin{align}
    \frac{\partial V_i}{\partial x_i} =
    \frac{1}{r^\alpha} + x_i\left( \frac{-\alpha}{r^{\alpha + 1}} \right)\left( \frac{x_i}{r} \right) =
    \frac{1}{r^{\alpha}} - \alpha \frac{x^2_i}{r^{\alpha + 2}}
    \tag{a}
\end{align}$

Ce qui donne :

$\begin{align}
    \text{div}_M \, \vec{V} =
    \frac{n}{r^{\alpha}} - \alpha \frac{\displaystyle\sum^n_{i=1}  x^2_i}{r^{\alpha + 2}} =
    \frac{n - \alpha}{r^\alpha} =
    \frac{n - \alpha}{\|\overrightarrow{OM}\|^\alpha}
    \tag{b}
\end{align}$

## 1.4 Laplacien

Si $f$ est une application de classe classe $\mathcal{C}^2$ sur $U$ à valeurs dans $\mathbb{R}$, on appelle *Laplacien* de $f$ en $M \in U$, le réel :

$\begin{align}
    \Delta_M \, f =
    \text{div}_M \, \left( \overrightarrow{\text{grad}} \, f \right)
    \tag{9}
\end{align}$

On a donc :

$\begin{align}
    \Delta_M \, f =
    \displaystyle\sum^n_{i = 1} \frac{\partial^2 f}{\partial x^2_i}(M)
    \tag{10}
\end{align}$

On note $\Delta \, f$ l'application $M \mapsto \Delta_M \, f$.

#### Propriétés

* L'application $f \mapsto \Delta \, f$ est un application linéaire de $\mathcal{C}^2 (U, \mathbb{R})$ dans $\mathcal{C}^0 (U, \mathbb{R})$.
* Elle vérifie, pour $f \in \mathcal{C}^2 (U, \mathbb{R})$ et $g \in \mathcal{C}^1 (U, \mathbb{R})$ :

$\begin{align}
    \Delta (f \, g) = f \, \Delta \, g + g \, \Delta \, f +
    2 \left( \overrightarrow{\text{grad}} \, f \right).\left( \overrightarrow{\text{grad}} \, g \right)
    \tag{11}
\end{align}$

*Exemple*

Cherchons les fonctions à symétrie sphérique, c'est-à-dire ne dépendant que de $r = \|\overrightarrow{OM}\|$, dont le laplacien est nul sur $U = \mathbb{R}^n \setminus \{ 0 \}$ (on dit alors qu'elles sont harmoniques).

**NB** > Cet exemple s'appuie sur les résultats de tous les exemples précédents pour produire son résultats.

Si $f(M) = \varphi(r)$, alors $\overrightarrow{\text{grad}}_M \, f = \varphi'(r).\vec{u} $ avec $\vec{u} = \displaystyle\frac{\overrightarrow{OM}}{r}$ et donc :

$\begin{align}
    \Delta_M (f) = \text{div}_M (\varphi'(r).\vec{u}) =
    \left( \overrightarrow{\text{grad}}_M \, \varphi'(r) \right).\vec{u} + \varphi'(r) \, \text{div}_M \, \vec{u}
    \tag{a}
\end{align}$

Soit :
$\begin{align}
    \Delta_M (f) =
    \varphi''(r).\vec{u}.\vec{u} + \varphi'(r) \displaystyle\frac{n - 1}{r} \\ =
    \varphi''(r) + \varphi'(r) \displaystyle\frac{n - 1}{r}
    \tag{b}
\end{align}$

Il s'agit donc de résoudre l'équation différentielle $r\varphi''(r) + (n - 1) \varphi'(r) = 0$.

## 1.5 Rotationnel

On suppose ici $n = 3$. Soient $\vec{V}$ une champ de vecteurs de classe $\mathcal{C}^1$ sur $U$, et $J$ sa matrice jacobienne.

La matrice $J - J^\top$ est une matrice antisymétrique, qui est donc de la forme :

$\begin{align}
    J - J^\top=
   \begin{pmatrix}
    0 & -r & q\\
    r & 0 & -p\\
    -q & p & 0\\
  \end{pmatrix} 
    \tag{12}
\end{align}$

C'est la matrice dans la base canonique de $\mathbb{R}^3$, de l'endomorphisme $\vec{u} \mapsto \vec{\omega} \times \vec{u}$ où $\vec{\omega}$ a pour composantes $(p, q, r)$.

**NB** Notation anglosaxonne du produit vectoriel (*cross product*).

<mark>Réviser le produit vectoriel</mark>

Ce vecteur $\vec{\omega}$ est appel *rotationnel* de $\vec{V}$ en $M$ et se note $\overrightarrow{\text{rot}}_M \, \vec{V}$.

On a donc :

$\begin{align}
  \overrightarrow{\text{rot}}_M \, \vec{V} =
  \left(
    \frac{\partial V_3}{\partial x_2} - \frac{\partial V_2}{\partial x_3}, \,
    \frac{\partial V_1}{\partial x_3} - \frac{\partial V_3}{\partial x_1}, \,
    \frac{\partial V_2}{\partial x_1} - \frac{\partial V_1}{\partial x_2}
  \right)
  \tag{13}
\end{align}$

#### Propriétés

* L'application $\vec{V} \mapsto \overrightarrow{\text{rot}} \, \vec{V}$ est un application linéaire de $\mathcal{C}^2 (U, \mathbb{R}^3)$ dans $\mathcal{C}^0 (U, \mathbb{R}^3)$.
* Elle vérifie, pour $f \in \mathcal{C}^1 (U, \mathbb{R})$ et $\vec{V} \in \mathcal{C}^1 (U, \mathbb{R}^3)$ :

$\begin{align}
    \overrightarrow{\text{rot}} (f \, \vec{V}) =
    f \, \overrightarrow{\text{rot}} \, \vec{V} +
    \overrightarrow{\text{grad}} \, f \times \vec{V}
    \tag{14}
\end{align}$

*Exemple*

Le rotationnel du champ de vecteurs $\vec{V}$ défini sur $\mathbb{R}^3$ par $\vec{V} = (x_3, x_1, x_2)$ est $\overrightarrow{\text{rot}} \, \vec{V} = (1, 1, 1)$.
