# Quick Start

```python
html_doc = """
<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
"""

from bs4 import BeautifulSoup
soup = BeautifulSoup(html_doc, 'html.parser')

print(soup.prettify())
```

```python
print(soup.title)
print(soup.title.name)
print(soup.title.string)
print(soup.title.parent.name)
print(soup.p)
print(soup.p['class'])
print(soup.a)
print(soup.find_all('a'))
print(soup.find(id='link3'))
```

```python
# Extraire tous les liens d'une page
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
# Extraire tout le texte d'une page
print(soup.get_text())
```

## Wikipédia : anciennes communes

Root : https://fr.wikipedia.org/wiki/Listes_des_anciennes_communes_de_France

```python
html_filename = 'Listes des anciennes communes de France — Wikipédia.html'
with open(html_filename, encoding='utf-8') as fp:
    soup = BeautifulSoup(fp)
    print(soup.prettify())
```

```python
# liste des liens vers les listes
html_filename = 'Listes des anciennes communes de France — Wikipédia.html'
with open(html_filename, encoding='utf-8') as fp:
    soup = BeautifulSoup(fp)
    links = soup.find_all('a', {'href': lambda x: x and x.startswith('/wiki/Liste_des_anciennes_communes_d')})
    for link in links:
        print(link.get('href'))
```

```python
# next step : recherches contextuelles : les liens qui sont enfants de telle ou telle section
# typiquement, ici, je cherche les liens enfants de la div class 'mw-parser-output', fille de la div id mw-content-text
# ensuite, il me faudra contextueliser cf. les frères : commune_1 > commune_2 par exemple, au sein d'un même li
import requests
url_ain = 'https://fr.wikipedia.org' + '/wiki/Liste_des_anciennes_communes_de_l%27Ain'
with requests.get(url_ain) as r:
    soup_ain = BeautifulSoup(r.content, 'lxml')
    #print(soup_ain.prettify())
    links = soup_ain.find_all('a', {'href': lambda x: x and x.startswith('/wiki/')})
    for link in links:
        print(link.get('href'))
```
