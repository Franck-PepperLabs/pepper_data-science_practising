## Incorporation with OpenCV

```python
import numpy as np
import cv2
r = np.ma.reshape(np.arange(256 * 256) % 256, (256, 256)) # 256x256 pixel array with a horizontal gradient from 0 to 255 for the red color channel
g = np.zeros_like(r)  # array of the same size and type as r but filled with 0s for the green channel
b = r.T   # transposed r will give a vertical gradient for the blue channel
ds = np.dstack([b, g, r])
cv2.imwrite('grandiens.png', ds) # OpenCV images are interpreted as BGR, the depth-stacked array will be written to an 8bit RGB PNG-file called 'gradient.png'
```

##### fromfunction, index n-d

```python
def f(i, j): return 10 * i + j
a = np.fromfunction(f, (5, 4), dtype=int); print(a)
a = np.fromfunction(lambda i, j: 10 * i + j, (5, 4), dtype=int); print(a)
a = np.array([[10 * i + j for j in range(4)] for i in range(5)]); print(a)

# pratique pour construire plein de matrices : ex. table de multiplication
n = 10
a = np.fromfunction(lambda i, j: (i + 1) * (j + 1), (n, n), dtype=int); print(a)
print(a[5 - 1, 6 - 1])
print(a[:5 - 1, 6 - 1])
print(a[3:7, 4:9])
print(a[::-2, ::3])
print(a[-2, ...]) # same as a[-2]
print(a[..., -2]) # same result, but it is the column
```

### Shape Manipulation


#### Changing the shape of an array

```python
a = np.floor(10 * rg.random((3, 4))); print(a)
print(a.shape)
# flatten
print(a.ravel())
b = a.reshape((6, 2)); print(b)
print(a.T)
print(a.T.shape)
# reshape and ravel return a new shape. resize is in place :
b.resize((4, 3)); print(b)
c = a.reshape(2, -1); print(c)
```

#### Stacking together different arrays

```python
a = np.floor(10 * rg.random((2, 2))); print(a)
b = np.floor(10 * rg.random((2, 2))); print(b)
c = np.vstack((a, b)); print(c)
c = np.hstack((a, b)); print(c)
a = np.arange(3); print(a)
b = a[::-1] * 2 + 3; print(b)
c = np.column_stack((a, b)); print(c) # diff from hstack
c = np.hstack((a, b)); print(c)
c = np.hstack((a.T, b.T)); print(c) # danger : x.T = x si x est 1-D => column_stack est incontournable... faux, cf. np.newaxis
c = c[:, np.newaxis]; print(c) # newaxis is an alias for None
a = np.arange(6); print(a)
b = a[:, None] * a; print(b) # nb > à la diff de pandas, * est bien le produit matriciel (@ dans Pandas)
# outer product
y = np.arange(3, 6); print(y)
z = np.outer(x, y); print(z)
z = x[:, np.newaxis] * y; print(z)
print(np.column_stack is np.hstack) # comportement différents
print(np.row_stack is np.vstack) # alias
a = np.arange(3) + 2
r = np.r_[a, 0, 1, a]; print(r)
c = np.c_[a, (10, 11, 12), a]; print(c)
```
